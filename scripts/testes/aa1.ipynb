```python
!pip install yfinance 
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go
import plotly.io as pio
import os
```

```python
commodities = {
    "Ouro": "GC=F",
    "Petróleo WTI": "CL=F",
    "Petróleo Brent": "BZ=F",
    "Gás Natural": "NG=F",
    "Milho": "ZC=F",
    "Soja": "ZS=F",
    "Trigo": "ZW=F",
    "Cobre": "HG=F",
    "Açúcar": "SB=F",
    "Café": "KC=F"
}
```

```python
start_date = "2020-01-01"
```

```python
for nome, ticker in commodities.items():
    print(f"Processando: {nome}")
    # 2. Baixa novos dados
    df = yf.download(ticker, start=start_date, interval="1d", progress=False)

    df.to_csv(f"{nome}.csv")
    print(f"Dados de {nome} salvos em: {f'{ticker}.csv'}")
```

```python
ouro = pd.read_csv("Ouro.csv")
petroleo_wti = pd.read_csv("Petróleo WTI.csv")
petroleo_brent = pd.read_csv("Petróleo Brent.csv")
gas_natural = pd.read_csv("Gás Natural.csv")
milho = pd.read_csv("Milho.csv")
soja = pd.read_csv("Soja.csv")
trigo = pd.read_csv("Trigo.csv")
cobre = pd.read_csv("Cobre.csv")
acucar = pd.read_csv("Açúcar.csv")
cafe = pd.read_csv("Café.csv")
```

```python
nomes = [
    "ouro", "petroleo_wti", "petroleo_brent", "gas_natural",
    "milho", "soja", "trigo", "cobre", "acucar", "cafe"
]

for nome, comm in zip(nomes, [ouro, petroleo_wti, petroleo_brent, gas_natural, milho, soja, trigo, cobre, acucar, cafe]):
    comm = comm.iloc[2:].copy()
    comm.rename(columns={"Price": "Date"}, inplace=True)
    comm.drop(columns=["Open", "High", "Low", "Volume"], inplace=True)
    comm.reset_index(drop=True, inplace=True)
    
    # Salva em CSV
    comm.to_csv(f"{nome}_limpo.csv", index=False)
```

```python
ouro_limpo = pd.read_csv("ouro_limpo.csv")    
petroleo_wti_limpo = pd.read_csv("petroleo_wti_limpo.csv")
petroleo_brent_limpo = pd.read_csv("petroleo_brent_limpo.csv")
gas_natural_limpo = pd.read_csv("gas_natural_limpo.csv")
milho_limpo = pd.read_csv("milho_limpo.csv")
soja_limpo = pd.read_csv("soja_limpo.csv")
trigo_limpo = pd.read_csv("trigo_limpo.csv")
cobre_limpo = pd.read_csv("cobre_limpo.csv")
acucar_limpo = pd.read_csv("acucar_limpo.csv")
cafe_limpo = pd.read_csv("cafe_limpo.csv")
```

```python
fig = go.Figure()

# Adiciona cada commodity ao gráfico
for nome, comm in zip(
    ["Ouro", "Petróleo WTI", "Petróleo Brent", "Gás Natural", "Milho", "Soja", "Trigo", "Cobre", "Açúcar", "Café"],
    [ouro_limpo, petroleo_wti_limpo, petroleo_brent_limpo, gas_natural_limpo, milho_limpo, soja_limpo, trigo_limpo, cobre_limpo, acucar_limpo, cafe_limpo]
):
    fig.add_trace(go.Scatter(x=comm['Date'], y=comm['Close'], mode='lines', name=nome))
```

```python
fig.update_layout(
    title="Preços de Commodities ao Longo do Tempo",
    xaxis_title="Data",
    yaxis_title="Preço de Fechamento",
    template="plotly_white"
)
```

```python

pio.write_html(fig, file="commodities.html", auto_open=True)
```

```python
dados = {
    "Ouro": pd.read_csv("ouro_limpo.csv", parse_dates=["Date"]),
    "Petróleo WTI": pd.read_csv("petroleo_wti_limpo.csv", parse_dates=["Date"]),
    "Petróleo Brent": pd.read_csv("petroleo_brent_limpo.csv", parse_dates=["Date"]),
    "Gás Natural": pd.read_csv("gas_natural_limpo.csv", parse_dates=["Date"]),
    "Milho": pd.read_csv("milho_limpo.csv", parse_dates=["Date"]),
    "Soja": pd.read_csv("soja_limpo.csv", parse_dates=["Date"]),
    "Trigo": pd.read_csv("trigo_limpo.csv", parse_dates=["Date"]),
    "Cobre": pd.read_csv("cobre_limpo.csv", parse_dates=["Date"]),
    "Açúcar": pd.read_csv("acucar_limpo.csv", parse_dates=["Date"]),
    "Café": pd.read_csv("cafe_limpo.csv", parse_dates=["Date"]),
}

# 2. Indexa cada série em 100 no primeiro valor
for nome, df in dados.items():
    primeiro = df["Close"].iloc[0]
    df["Indexado"] = df["Close"] / primeiro * 100

# 3. Monta o gráfico usando a coluna "Indexado"
fig = go.Figure()
for nome, df in dados.items():
    fig.add_trace(
        go.Scatter(
            x=df["Date"],
            y=df["Indexado"],
            mode="lines",
            name=nome
        )
    )
```
